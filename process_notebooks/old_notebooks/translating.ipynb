# Translate Japanese text into English using googletrans
Kamon descriptions and subcategory titles are in Japanese and should be translated into english. I'm going to use the googletrans python library to do so.


```python
from googletrans import Translator
import pandas as pd
import asyncio
import nest_asyncio
```

```python
# read Kamon df
df = pd.read_csv('Data/kamon_data.csv')
df.head()
```

```python
#Translate Kamon descriptions and subcategory titles from Japanese to English
translator = Translator()

async def translate_text(text):
    try:
        result = await translator.translate(text, src='ja', dest='en')
        return result.text
    except Exception as e:
        return None
    
text = df.iloc[0]["description"]
translated = await translate_text(text)
print(translated)

```

Got async translate to work, now testing below on description. Also using nest_asyncio. Instead of applying the function to each row, translate entire column and assign the result.

```python
async def translate_column(texts):
    return [await translate_text(x) for x in texts]

nest_asyncio.apply()
loop = asyncio.get_event_loop()
df["English_Description"] = loop.run_until_complete(translate_column(df["description"]))
```

```python
# HOLY SHIT SAVE THE RESULT NOW IT TOOK 37 MINUTES TO RUN
df.to_csv('Data/kamon_engdesc_only.csv', index=False)
```

```python
#Get all unique subcategory titles from dataframe
unqiue_subtitles = df["subcategory_title"].unique()
unqiue_subtitles
```

```python
#now translate this array
english_subtitles = loop.run_until_complete(translate_column(unqiue_subtitles))
```

```python
english_subtitles
```

```python
#remove all / characters
english_subtitles_cleaned = [s.replace(' / ', '') if isinstance(s, str) else s for s in english_subtitles]
english_subtitles_cleaned
```

```python
# put this after the next cell
df['English_Subcategory'] = df['subcategory_title'].map(dict(zip(unqiue_subtitles, english_subtitles_cleaned)))
```

```python
#final save
df.to_csv('Data/kamon_english.csv', index=False)
```
